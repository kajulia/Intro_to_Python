Oblicz średnią z listy zawierającej liczby bezpośrodednio z użyciem pętli (bez stosowania funkcji np. sumowania).
Sprawdzić zachowanie gdy lista jest pusta. 

```python
#create list
list1 = random.sample(range(0, 101), 10)
print(list1)


#average-counting loop
sum1, total1 = 0, 0
for i in list1:
    sum1 += i
    total1 += 1
av1 = sum1/total1
print(av1)
```

Dla chętnych obliczanie wariancji albo standardowego odchylenia.

```python
#create list
list3 = random.sample(range(0, 101), 10)
print(list3)

#average-counting loop
sum3, total3 = 0, 0
for i in list3:
    sum3 += i
    total3 += 1
av3 = sum3/total3
print(av3)

#variance-counting loop
v3 = 0
for i in list3:
    v3 += ((i-av3)**2)
var3 = v3/total3
print(var3)

#standard deviation
sd3 = math.sqrt(var3)
print(sd3)


    
```

max z listy (ale podobnie jak średnia)

```python
#create list
list6 = random.sample(range(0, 101), 10)
print(list6)


#the-biggest loop
max6 = 0
for i in list6:
    if max6 > i:
        max6 = max6
    else:
        max6 = i

print(max6)
```

```python
#create list
list7 = random.sample(range(0, 101), 10)
print(list7)

#the-second list
max7 = list7[0]
sec7 = 0
for i in list7:      
        if i > max7:  
            sec7 = max7 
            max7 = i  
        elif sec7 < i:  
            sec7 = i  

print (sec7, max7)
        
```
